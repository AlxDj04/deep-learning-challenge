```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
app_new = application_df.drop(columns=['EIN', 'NAME'])
app_new.head()
```

```python
u_counts = app_new.nunique()
u_counts
```

```python
app_new['APPLICATION_TYPE'].value_counts()
```

```python
cutoff_value = 200
application_type_counts = app_new['APPLICATION_TYPE'].value_counts()
application_types_to_replace = application_type_counts[application_type_counts < cutoff_value].index.tolist()
# Replace in dataframe
for app in application_types_to_replace:
    app_new['APPLICATION_TYPE'] = app_new['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
app_new['APPLICATION_TYPE'].value_counts()
```

```python
app_new['CLASSIFICATION'].value_counts()
```

```python
cutoff_value = 200
classifications_type_counts = app_new['CLASSIFICATION'].value_counts()
classifications_to_replace = classifications_type_counts[classifications_type_counts < cutoff_value].index.tolist()
# Replace in dataframe
for cls in classifications_to_replace:
    app_new['CLASSIFICATION'] = app_new['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
app_new['CLASSIFICATION'].value_counts()
```

```python
df_encoded = pd.get_dummies(app_new)
df_encoded.head()
```

```python
X = df_encoded.drop(columns=['IS_SUCCESSFUL', 'INCOME_AMT_50M+','INCOME_AMT_5M-10M', 'SPECIAL_CONSIDERATIONS_Y', 'SPECIAL_CONSIDERATIONS_N'], axis=1)
y = df_encoded['IS_SUCCESSFUL']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1)
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units=100, input_dim=41, activation='relu'))
nn.add(tf.keras.layers.Dense(units=30, activation='sigmoid'))
nn.add(tf.keras.layers.Dense(units=30, activation='linear'))
nn.add(tf.keras.layers.Dense(units=30, activation='tanh'))

nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# Check the structure of the model
nn.summary()
```

```python
nn.compile(
    loss = 'binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
nn.fit(X_train_scaled, y_train, epochs=20)
#nn.fit(X_train_scaled, y_train, epochs=80)
#nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
